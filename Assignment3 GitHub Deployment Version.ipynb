## **Data Introduction & Background**

For Assignment 3, I'm looking at employment-based green card sponsorship in the U.S., using PERM data the Department of Labor released for fiscal year 2025.

PERM stands for Program Electronic Review Management, the first real step in most employment-based green card cases. It boils down to the employer proving to the government they tried to hire a qualified U.S. worker and couldn't find one, which is what opens the door for a foreign worker to fill the role instead.

I picked this topic because me and a lot of my friends are going through the process right now. :)

It's gotten harder to get a green card through employment sponsorship lately, for a few reasons. Fewer employers want to take on this lengthy, unpredictable process in the first place. The wage floor the Department of Labor sets for each job function climbs every year, raising the bar for what a foreign worker needs to be offered. Filing fees, admin costs, and rejection/RFE rates are all up too, scaring off employers who might otherwise take the risk.

So for this project, I wanted to dig into the real PERM data myself. Questions I kept coming back to:

- How does the wage offered to sponsor someone differ by company size?
- Do wages differ much across occupation types?
- Which job titles show up most in PERM filings, and where geographically?
- Does a company's age relate to what it pays, or how often it sponsors at all?
- Does processing time relate to whether a case gets certified or denied?

Getting approved depends on more than any one factor, education, wage, job type, and required experience all play in. But the data still gives a decent window into these patterns.

**What's in the data:** this public disclosure dataset has 137 columns and 147,056 rows. Each row is one real PERM application filed by an employer sponsoring a foreign worker for an employment-based green card during Fiscal Year 2025 (10/1/2024 through 9/30/2025). Each row covers the employer (business name, industry code, number of employees, year established), the attorney or agent representing them if there is one, the job itself (title, SOC occupation code, offered wage, occupation type), the prevailing wage determination number, the worksite location (city, county, state), details on how the employer recruited (job order dates, ads placed, methods used), and how the case turned out (status, dates received and decided).

**Data source:** [DOL Office of Foreign Labor Certification](https://www.dol.gov/agencies/eta/foreign-labor/performance)

<a id="technique-1"></a>
#### **1. Box Plots: Green Card Applicant's Wage Distribution by Company Size**

In this case, we're using box plots to compare how wage distributions differ across different company sizes. Box plots work well here because we're interested in comparing distributions across categories, not just averages.

Using box plots lets us read off center, spread, skew, and any outliers side by side across all six groups at a glance:

- `Center line`: the median wage for each group
- `Box height`: the spread of the middle 50% of wages
- `Single dots` above/below the whiskers: any outliers with unusually high or low wage offers
- `Overall shape`: whether wages are tightly clustered or widely spread, and how many extreme outliers each group contains

Using box plots as the visualization technique here makes it easy to answer questions like:
- Do larger companies generally pay more to sponsor a foreign worker for a Green Card?
- How consistent are wages within each company size category?

Compared to other visualization techniques, like a bar chart which only compares averages, box plots reveal differences in both wage level and wage variability, which meets our needs better here.

<a id="technique-2"></a>
#### **2. Violin Plots: Green Card Applicant's Wage Distribution by Company Age**

A violin plot sombines a box plot with with a mirrored density curve. It allows us to compare both summary statistics and the overall shape of the wage distribution. In this visualization, the embedded box plot shows the median and interquartile range, while the surrounding density curve highlights where wages are most concentrated.

- `Width at any point`: how many applicants fall at that wage level. Wider sections indicate higher density.
- `Embedded box plots`: the median and spread of the middle 50% of wages.
- `Overall silhouette`: whether wages cluster tightly around a single value, spread evenly, or have multiple common wage points (a violin with more than one peak)

Using a violin plot here helps answer questions like:
- Do older companies pay more consistently than newer ones, or just differently on average?
- Does each age group have a single common wage range, or are there multiple wage clusters?
- How does the overall wage distribution change as companies age??

Compared to a box plot, which only shows summary statistics (median, quartiles, outliers), a violin plot reveals the full density shape of the distribution. Since the previous visualization focuses on box plots, using a violin plot here provides additional insight into how wage distributions differ across company age groups.

<a id="technique-3"></a>
#### **3. Histogram: PERM Wage Distribution by Occupation Type (Professional Occupation vs. Non-Professional Occupation)**

A stacked histogram works well to answer this question, because we want to learn both the wage distribution and the occupation breakdown at the same time. A stacked histogram shows:

- `Overall bar height`: the total number of applications at that wage level, across all occupation types combined
- `Colored segments within each bar`: how much of that total comes from each occupation type (Professional occupation, Non-professional, or College/University Teacher)

Using a stacked histogram here makes it easy to answer questions like:
- Where does the bulk of PERM wage offers fall, regardless of occupation type?
- Do professional occupations dominate over the others?

Unlike the box and violin plots used in Plots 1 and 2, which show an overall summary-statistics view, a histogram represents the full and granular shape of the distribution using raw record counts. It shows exactly how many applicants fall into each wage bracket. This makes it a genuinely different technique from the previous two charts.

<a id="technique-4"></a>
#### **4. Bar Chart: Top 10 Employment-Based Green Card Applicant Job Titles**

A bar chart is one of the most straightforward visualization techniques for this kind of question. I picked a bar chart to answer "top 10..." type questions because it matches our need best: when the goal is simply to rank categories by count, a bar chart communicate the most directly with its bar length:

- `Bar length`: directly shows how many applicants applied for a green card under that job title, making it immediately clear which roles appear most often in the dataset
- `Horizontal orientation`: makes even very long job title labels easy to read without rotating or truncating text
- `Sorted order`: ranking bars from longest to shortest turns the chart into an instant leaderboard

Using a bar chart here makes it easy to answer questions like:
- Which job titles see the most PERM sponsorship activity?

Overall, a bar chart is a straightforward ranking of counts across a categorical variable.

<a id="technique-5"></a>
#### **5. Tree Map: PERM Employer Location by City (Top 20)**

Tree maps show hierarchical data as nested rectangles. I picked this one partly because it was covered in this week's Module 3. Each rectangle size maps to volume (number of
applications), and color here is just used to tell one city apart from the next, it's not tied to a second metric.

This works well for a "where's the activity concentrated" kind of question:
- `Size of each rectangle`: the bigger the box, the more applications came from that city
- Comparing 20 cities side by side is a lot easier here than scrolling through a ranked list
- State isn't shown as a separate layer since city names alone don't need it. But I added it in the hover tooltip instead, so the chart doesn't get crowded with an extra level of nesting

<a id="technique-6"></a>
#### **6. Strip Plot: Processing Days by Case Outcome**

A strip plot fits here because we're comparing a continuous variable, processing days, against a categorical one, case outcome, the same basic comparison as the box and violin plots back in Plots 1 and 2. But this time every single case shows up as its own dot instead of getting folded into a median or a quartile.

- `Each dot`: one PERM application, plotted by how many days it took from received date to decision
- `Jitter`: spreads the dots left and right inside each category so they do not just stack into a straight line, which is what actually lets you see density instead of a solid bar
- `Lower opacity`: with tens of thousands of points per group, dropping the opacity way down helps too, denser spots come out darker and sparse spots almost disappear, so the eye can still find where cases really cluster

This setup makes it easy to answer questions like:
- Do certified and denied cases take about the same time, or does one resolve faster?
- Does an outcome hide subgroups, like a batch of unusually fast denials, that a median would smooth over?

Compared to the box and violin plots in Plots 1 and 2, a strip plot keeps every point visible, a real tradeoff since at this sample size it needs jitter and low opacity just to stay readable, but nothing gets averaged away either.

### **1. Plotly: for Building the Plots**

- **Website:** https://plotly.com/python/
- **Developer:** Plotly Technologies Inc. (Montreal-based)<br>
- **Open Source:** Yes, MIT license<br>
- **Free to use?** Yes, Plotly for Python is free and open source<br>
- **Integrates with Jupyter?** Yes, works natively, calling `.show()` renders each figure as an interactive widget right in the output cell<br>
- **Declarative or procedural?** Declarative. You write code describing what the final result should look like, not step-by-step instructions on how to draw it. You specify the data, the x/y mapping, the grouping variable, and Plotly figures out how to actually draw it, which is different from Matplotlib's more step-by-step drawing commands. That's what makes rebuilding six charts on every dropdown change manageable, the update function just calls something like `px.box(filtered_df, x="company_size_bin", y="annual_wage")` again and gets a complete, correctly formatted chart back, no manual redraw logic needed.<br>

#### **Reference**
https://plotly.com/python/getting-started/<br>

#### **Installation**

- Install with pip: `pip install plotly`
- or conda: `conda install -c conda-forge plotly`

#### **Why Plotly?**

**Two main reasons I picked Plotly Express here are:**

1. it covers every chart type I needed, box plots, violin plots, histograms, bar charts, treemaps, strip plots.
2. it's genuinely easy to use, no complicated custom code, and the output looks good by default, hover details, color palettes that look more polished than Matplotlib out of the box.

**A few other reasons it made sense:**

- **Low code effort for good-looking output**, a lot comes built in
- **Renders inline in Jupyter** with no real setup
- **Hover, zoom, pan all work by default**, no extra config
- **Pairs cleanly with `ipywidgets`** for the linked filtering in this dashboard, each chart just gets rebuilt inside the update function on the filtered data
- **Documentation and example gallery are solid**, useful when a specific chart type wasn't obvious how to build

#### **Some Limitations I Ran Into**

- **Large datasets get dense fast.** The strip plot in Plot 6 has tens of thousands of points, and the defaults made it unreadable, had to manually tune marker opacity and jitter to get any real density visible instead of a solid blob
- **Fine layout control needs workarounds.** Things like the multi-line x-axis labels with sample counts in Plots 1 and 2 aren't a built-in option, had to manually generate tick labels instead of using the high-level API directly
- **No built-in dashboard state management.** Plotly Express only handles individual charts. Linking three dropdowns to six charts at once had to be built separately with `ipywidgets`, Plotly alone doesn't coordinate multiple charts
- **Performance can slow down with large datasets.** A lot of people online list this as a weakness, but with roughly 147,056 rows by 15 columns after cleaning, I didn't notice much of a slowdown myself. My notebook crashed a few times, but rendering the actual visualizations stayed quick

**Other reasons...**

- **Only works with a live Python kernel.** The interactivity doesn't survive if the notebook gets exported to static HTML or viewed on GitHub, it only works while actually running, which is exactly why this dashboard needed to be deployed on Colab and not just committed as a file

First, let's take a look at the `dataframe shape`, `head rows`, and `missing values` in this dataset. (Note: The DOL offers an official data layout explanation for each column [here](https://www.dol.gov/sites/dolgov/files/ETA/oflc/pdfs/PW_Record_Layout_FY2025_Q4.pdf))

```python
### GITHUB NOTE
# Commented out for performance, and so this notebook can run locally after being
# downloaded from GitHub. The full EDA process can be seen in the Vocareum notebook
# or the Google Colab version.

# check shape
# df.shape

# Output: (147056, 137)

# df.head(3)
```

**Data Cleaning Process**

1. **Dropped all columns with > 90% missing values**: 27 columns gone, taking the dataframe from 137 columns down to 110.
2. **Dropped columns not needed for visualization**, another 95 gone, down from 110 to 15.
3. **Calculated and normalized the new fields the plots need**:
   - **Normalize annual wage:** the raw `JOB_OPP_WAGE_FROM` field wasn't on one consistent unit, some rows were `Year`, others `Hour`, `Month`, `Bi-Weekly`, or `Week`. Converted everything to an annualized number so wages could actually be compared.
   - **Company size bins**, based on `EMP_NUM_PAYROLL`: `<50`, `50-200`, `200-500`, `500-1000`, `1000-5000`, `5000+`.
   - **Company age**, calculated as 2025 - `EMP_YEAR_COMMENCED`, then binned into `<5 years`, `5-20 years`, `20-50 years`, `50+ years`.
   - **Processing days**, just `DECISION_DATE` - `RECEIVED_DATE`.

```python
# PLOT 1
# box plot: company size vs. wage

# calculate sample size per company size bin
group_counts_1 = (
    df_clean.dropna(subset=["company_size_bin", "annual_wage"])
    .groupby("company_size_bin", observed=True)
    .size()
)

# build x axis labels to include the count
label_map_1 = {
    bin_name: f"{bin_name}<br>(n={count:,})"
    for bin_name, count in group_counts_1.items()
}

fig1 = px.box(
    df_clean.dropna(subset=["company_size_bin", "annual_wage"]),
    x="company_size_bin",
    y="annual_wage",
    category_orders={"company_size_bin": ["<50", "50-200", "200-500", "500-1000", "1000-5000", "5000+"]},
    title="<b>PERM Wage Distribution by Company Size (FY2025)</b><br><sup>Do larger companies offer higher sponsorship wages to foreign workers?</sup>",
    labels={
        "company_size_bin": "Company Size (# of Employees)",
        "annual_wage": "Annual Wage (USD)"
    }
)

# apply new count labels to the x-axis
fig1.update_xaxes(
    tickmode="array",
    tickvals=[str(v) for v in group_counts_1.index],
    ticktext=[label_map_1[bin_name] for bin_name in group_counts_1.index]
)

fig1.update_layout(
    height=600,
    autosize=True,
    title_x=0.5,
    yaxis_tickformat="$,.0f"
)

overall_median = df_clean["annual_wage"].median()
fig1.add_hline(
    y=overall_median,
    line_dash="dash",
    line_color="red"
)
fig1.add_annotation(
    text=f"<span style='color:red'>┅</span> Overall median wage: ${overall_median:,.0f}",
    xref="paper", yref="paper",
    x=0.5, y=1.06,
    showarrow=False,
    font=dict(size=13, color="gray"),
    align="left"
)

fig1.show(config={"responsive": True})
```

From the box plots, we can tell that larger companies tend to offer higher and more consistent wages when sponsoring foreign workers for employment-based green cards. Besides, the plot reveals the companies with <50 employees actually account for the largest share of PERM applications (n=38,077), it's more than any other group, suggesting that most PERM sponsorship activity comes not from large employers, but from a high volume of small companies.

```python
from ipywidgets import interact, Dropdown
from IPython.display import clear_output, display, Markdown
import plotly.express as px

# build dropdown options from the cleaned data
state_options = ["All"] + sorted(df_clean["PRIMARY_WORKSITE_STATE"].dropna().unique().tolist())
size_options = ["All", "<50", "50-200", "200-500", "500-1000", "1000-5000", "5000+"]
job_options = ["All"] + sorted(df_clean["PWD_SOC_TITLE"].dropna().unique().tolist())


def update_dashboard(state, size, job):
    clear_output(wait=True)
    display(Markdown(f"Dashboard updating in progress..."))

    # filter data based on current dropdown selections
    filtered_df = df_clean.copy()
    if state != "All":
        filtered_df = filtered_df[filtered_df["PRIMARY_WORKSITE_STATE"] == state]
    if size != "All":
        filtered_df = filtered_df[filtered_df["company_size_bin"] == size]
    if job != "All":
        filtered_df = filtered_df[filtered_df["PWD_SOC_TITLE"] == job]

    clear_output(wait=True)
    display(Markdown(f"---\n\n✅ **Dashboard Updated**: Showing **{len(filtered_df):,} applications** matching current filters"))

    # PLOT 1: box plot: company size vs. wage
    group_counts_1 = (
        filtered_df.dropna(subset=["company_size_bin", "annual_wage"])
        .groupby("company_size_bin", observed=True)
        .size()
    )
    label_map_1 = {
        bin_name: f"{bin_name}<br>(n={count:,})"
        for bin_name, count in group_counts_1.items()
    }
    fig1 = px.box(
        filtered_df.dropna(subset=["company_size_bin", "annual_wage"]),
        x="company_size_bin",
        y="annual_wage",
        category_orders={"company_size_bin": ["<50", "50-200", "200-500", "500-1000", "1000-5000", "5000+"]},
        title="<b>PERM Wage Distribution by Company Size (FY2025)</b><br><sup>Do larger companies offer higher sponsorship wages?</sup>",
        labels={
            "company_size_bin": "Company Size (# of Employees)",
            "annual_wage": "Annual Wage (USD)"
        }
    )
    fig1.update_xaxes(
        tickmode="array",
        tickvals=list(group_counts_1.index),
        ticktext=[label_map_1[bin_name] for bin_name in group_counts_1.index]
    )
    fig1.update_layout(
        height=600,
        title_x=0.5,
        yaxis_tickformat="$,.0f"
    )
    fig1.show()

    # PLOT 2: violin plot: company age bin vs wage
    group_counts_2 = (
        filtered_df.dropna(subset=["company_age_bin", "annual_wage"])
        .groupby("company_age_bin", observed=True)
        .size()
    )
    label_map_2 = {
        bin_name: f"{bin_name}<br>(n={count:,})"
        for bin_name, count in group_counts_2.items()
    }
    fig2 = px.violin(
        filtered_df.dropna(subset=["company_age_bin", "annual_wage"]),
        x="company_age_bin",
        y="annual_wage",
        box=True,
        points=False,
        category_orders={"company_age_bin": ["<5 years", "5-20 years", "20-50 years", "50+ years"]},
        title="<b>PERM Wage Distribution by Company Age (FY2025)</b><br><sup>Are older, more established companies more likely to sponsor foreign workers?</sup>",
        labels={
            "company_age_bin": "Company Age (Years in Business)",
            "annual_wage": "Annual Wage (USD)"
        }
    )
    fig2.update_xaxes(
        tickmode="array",
        tickvals=list(group_counts_2.index),
        ticktext=[label_map_2[bin_name] for bin_name in group_counts_2.index]
    )
    fig2.update_layout(
        height=600,
        title_x=0.5,
        yaxis_tickformat="$,.0f"
    )
    fig2.show()

    # PLOT 3: histogram: wage distribution by occupation type
    fig3 = px.histogram(
        filtered_df.dropna(subset=["annual_wage", "OCCUPATION_TYPE"]),
        x="annual_wage",
        color="OCCUPATION_TYPE",
        title="<b>PERM Wage Distribution by Occupation Type (FY2025)</b><br><sup>Do professional occupations command higher sponsorship wages?</sup>",
        labels={"annual_wage": "Annual Wage (USD)"},
        category_orders={"OCCUPATION_TYPE": ["Professional occupation", "Non-professional", "College/University Teacher"]}
    )
    fig3.update_traces(xbins=dict(start=0, end=1000000, size=40000))
    fig3.update_layout(
        height=600,
        title_x=0.5,
        xaxis_tickformat="$,.0f",
        bargap=0.1,
        xaxis=dict(tickmode="linear", dtick=40000, tickangle=45),
        yaxis_title="Number of Applications"
    )
    fig3.show()

    # PLOT 4: bar chart: top 10 job titles
    top_titles = filtered_df["JOB_TITLE"].value_counts().head(10).reset_index()
    top_titles.columns = ["JOB_TITLE", "count"]
    fig4 = px.bar(
        top_titles,
        x="count",
        y="JOB_TITLE",
        orientation="h",
        title="<b>Top 10 Employment-Based Green Card Applicant Job Titles</b><br><sup>Which jobs are most likely to get sponsored?</sup>",
        labels={"JOB_TITLE": "Job Title"}
    )
    fig4.update_layout(
        yaxis={"categoryorder": "total ascending"},
        title_x=0.5,
        height=600,
        xaxis_title="Number of Applications",
    )
    fig4.show()

    # PLOT 5: treemap: PERM job city distribution
    top_cities = (
        filtered_df.dropna(subset=["PRIMARY_WORKSITE_STATE", "city_clean"])
        .groupby(["PRIMARY_WORKSITE_STATE", "city_clean"])
        .size()
        .reset_index(name="count")
        .nlargest(20, "count")
    )
    fig5 = px.treemap(
        top_cities,
        path=["city_clean"],
        values="count",
        custom_data=["PRIMARY_WORKSITE_STATE"],
        title="<b>PERM Employer Location by City (Top 20)</b><br><sup>Where are you most likely to find a green-card-sponsoring employer?</sup>",
        labels={"city_clean": "City", "count": "Number of Applications"}
    )
    fig5.update_traces(
        hovertemplate="<b>%{label}</b><br>State: %{customdata[0]}<br>Applications: %{value}<extra></extra>"
    )
    fig5.update_layout(
        title_x=0.5,
        height=600
    )
    fig5.show()

    # PLOT 6: strip plot: processing days by case outcome
    plot6_df = filtered_df[
        filtered_df["certification_group"].isin(["Certified", "Denied"])
    ].dropna(subset=["processing_days"])

    fig6 = px.strip(
        plot6_df,
        x="certification_group",
        y="processing_days",
        color="certification_group",
        title="<b>Processing Days by Case Outcome (FY2025)</b><br><sup>Does how long an application takes relate to whether it gets approved?</sup>",
        labels={"certification_group": "Case Status", "processing_days": "Processing Days"}
    )
    fig6.update_traces(
        marker=dict(size=3, opacity=0.55),
        jitter=0.8
    )
    fig6.update_layout(
        height=600,
        title_x=0.5
    )
    fig6.show()


# connect the dropdowns to the update function
label_style = {"description_width": "120px", "font_weight": "bold"}

state_dropdown = Dropdown(
    options=state_options,
    description="State:",
    style=label_style,
    layout={"width": "250px"}
)
size_dropdown = Dropdown(
    options=size_options,
    description="Company Size:",
    style=label_style,
    layout={"width": "250px"}
)
job_dropdown = Dropdown(
    options=job_options,
    description="Job Function:",
    style=label_style,
    layout={"width": "500px"}
)

interact(
    update_dashboard,
    state=state_dropdown,
    size=size_dropdown,
    job=job_dropdown
)
```

## **Problems Met & Reflection**

<a id="reflection-1"></a>
### **1. 3 hours of debugging later, it turned out to be a rendering race condition, not a Google Colab or Plotly bug**

It took about 3 hours to debug why the final dashboard wouldn't render properly in Google Colab, and it turned out to be one of the trickier technical problems in this project, honestly more interesting than I expected.

Each of the 6 plots rendered fine when tested individually, but once combined into the final linked dashboard, charts would flicker, disappear, or only partially render. The pattern was inconsistent at first, sometimes only the first plot showed, sometimes only the last one, sometimes plots would appear and then vanish the moment the next one started rendering.

I tried a few fixes, roughly in this order:
- `google.colab.output.enable_custom_widget_manager()`, to properly connect Plotly's rendering to Colab's widget system
- `output.no_vertical_scroll()`, to stop Colab from clipping the output inside a fixed-height scrollable container
- `time.sleep(1.5)` between each `fig.show()` call, to give the browser time to finish rendering one chart before starting the next

Each helped a little, none fully solved it. The real clue came from watching how the charts failed, whichever chart rendered last tended to stay stable, and everything rendered before it would disappear. That pointed to the actual cause: all six figures were getting pushed into the same shared output stream one after another, and each new `.show()` call forced Colab to recompute and redraw the entire output area, wiping out whatever had rendered just before it. Adding delays between charts didn't fix this, it just changed when the wipeout happened.

The fix was to stop sharing one output stream entirely. I created six separate `ipywidgets.Output()` containers, one per chart, so each figure renders into its own isolated space instead of competing for the same one. Combined with switching from `interact()` to manually binding `.observe()` on each dropdown, this finally gave a stable dashboard, no more flickering, no more charts disappearing on scroll.

**Takeaway**: the bug wasn't really about Plotly, Colab, or ipywidgets individually, it was a race condition in how repeated output calls competed for the same rendering space. Isolating each chart into its own container removed the competition instead of trying to manage the timing around it.

<a id="reflection-2"></a>
### **2. Finding real, reliable individual-level data source is harder than picking a dataset off Kaggle**

Finding quality data, especially individual-level records, turned out to be a real pain. I could've just grabbed something off Kaggle, but I wanted first-hand data from a reliable source for this one, and that was way harder than expected. A few topics I tried before landing here:

- **Air Quality Data:** with the recent Canadian wildfire smoke, I got curious about Michigan's air quality history. Found something from the [EPA](https://aqs.epa.gov/aqsweb/airdata/download_files.html), but it only had a handful of columns, lat/long and air quality readings, not enough for what this assignment needed. Getting more would've meant stitching together multiple sources, and I wasn't even sure the extra data existed publicly. Gave up on this one.
- **Ann Arbor Parking Violations:** the city has a public data portal, but pulling actual ticketing data (location, time, car model) requires a formal request that can take weeks to get approved. Not happening on this timeline.
- **Farmer Suicide Rates:** remembered a news story on this, saying farmer's suicide rate is extremely high in the U.S., so wanted to check if raw data backed it up. Found [CDC WONDER](https://wonder.cdc.gov/controller/datarequest/D76), a public hub with suicide data, but federal law only lets them release aggregate summaries, no individual records. Dead end.

Long story short, finding good, individual-level, publicly available data is nowhere near as easy as it sounds. :(

<a id="reflection-3"></a>
### **3. Cleaning real-world data means fixing entry mistakes, not just dropping them**

Even in something as official as a green card filing, people mess up. This dataset was cleaner than most I looked at, but a few things still needed fixing.

Plot 1 (Company Size vs. Wage) is the clearest example. After plotting the raw numbers, some wages showed up over $500,000,000, obviously not real. Pulling the top outliers with `df_clean.nlargest(20, "annual_wage")[["JOB_OPP_WAGE_FROM", "JOB_OPP_WAGE_PER", "annual_wage", "JOB_TITLE"]]` showed most of them were senior or manager titles.

My best guess: someone meant to file an annual salary but picked "Hour" as the pay unit by mistake. Run that through my annualizing function and a normal \$150k salary turns into \$150k x 2,080 hours, well over \$300 million. Instead of tossing those rows, since the underlying data was still good, I wrote a separate function to catch and fix that specific pattern. A few smaller formatting issues popped up too, nothing major, overall the data held up well.

The six charts couldn't just be picked at random, they all had to actually respond to the same three controls, State, Company Size, Job Function. I had other chart ideas and side topics I wanted to dig into, but once I built them out, some just didn't play nice with the shared controls, so they got cut.
